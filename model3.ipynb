```python
import numpy as np
from keras.models import model_from_json

path = "C:/Users/IIT JAMMU/Documents/Flipkart/"

# Step 1: Load model architecture from JSON file
json_file_path = path + 'model.json'
with open(json_file_path, 'r') as json_file:
    loaded_model_json = json_file.read()

loaded_model = model_from_json(loaded_model_json)

# Step 2: Load model weights from binary file (.bin)
weight_bin_file_path = path + 'model.weights.bin'

# Load weights from the binary file into a numpy array
weights = np.fromfile(weight_bin_file_path, dtype=np.float32)

# Set the loaded weights to the model's layers
layer_index = 0
for layer in loaded_model.layers:
    layer_weights = weights[layer_index : layer_index + np.prod(layer.weights[0].shape)]
    layer.set_weights([layer_weights])
    layer_index += np.prod(layer.weights[0].shape)

# Now you have your model loaded with architecture and weights
```
